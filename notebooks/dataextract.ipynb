Meet the penguins:
The palmerpenguins data contains size measurements for three penguin species observed on three islands in the Palmer Archipelago, Antarctica.

Goal is to produce a model able to classify the penguins into three different species.

The second file is penguins_Iter.csv 
Lets combine two given sets in one data file that will be used further for classification model 

```python
import pandas as pd

df= pd.read_csv('penguins_lter.csv')
#Replace any null values 
df = df.fillna(0)
df.tail()
```

```python
# Lets add the columns below to a new dataframe 
# The ratio of isotope Nitrogen (15) and Carbon (13)
columns = ['Delta 15 N (o/oo)','Delta 13 C (o/oo)']
new_df = pd.DataFrame()
for i in columns:
    new_df[i] = df[i]
new_df.head()
```

```python
#Reading the second dataset for this data
df = pd.read_csv('penguins_size.csv')
df = df.fillna(0)
df.head()
```

```python
#concatenating the two dataframes
df = pd.concat([df, new_df], axis=1)
labels = df['species'].unique()
df.head()
```

```python
# Let see which species are there, there are 3 species 
labels = list(labels)
labels
```

```python
# Encoding target variables 
from sklearn.preprocessing import LabelEncoder 
le = LabelEncoder()
df['species'] = le.fit_transform(df['species'])
encodes = df['species'].unique()
encodes = list(encodes)
```

```python
# Species are encoded as 0,1,2 = (Adelie, Chinstrap, Gentoo)
encodes
```

```python
# create a txt files with encoding 
with open('label-encodes.txt', 'w') as f:
    for l, e in zip(labels, encodes):
        e = str(e)
        f.write(l)
        f.write(',')
        f.write(e)
        f.write('\n')
```

```python
# Finally writing a new file to work with in classification model 
df.to_csv('../notebooks/data.csv')
```
